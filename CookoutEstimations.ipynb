```python
# Faith Nchang
import math


PRICE_BURGER_PACKAGE = 24
BURGERS_PER_PACKAGE = 12

PRICE_PATTIES_PACKAGE = 12
PATTIES_PER_PACKAGE = 3

PRICE_BUNS_PACKAGE = 6
BUNS_PER_PACKAGE = 12

PRICE_PER_SODA_CAN = 1.5

PRICE_PER_CHIP_BAGS = 1

HOSTING_PRICE_PER_HOUR = 75
MIN_HOURS = 3


while True:
    print("********************** COOKOUT ESTIMATIONS *********************************************")
    print("Welcome to Faith's Cookouts. Let's estimate your cookout cost.")
    name = input("What is your name: ")
    # prompts the user for an estimated number of guests the customer will have
    guests = input(f"Hi {name}! what is the total number of guests who will be attending the cookout: ")
    # Ensures the number of guests is a whole number >=1
    while guests.isdigit() == False or int(guests) < 1:
        guests = input("Please enter a whole number above 0: ")

    # displays the food choices
    print()
    print("  You have 3 food options to select one to serve your guests")
    print("\t1. Burgers")
    print("\t2. Chicken Sandwich")
    print("\t3. Meatball Subs")
    print()

    # the customer will have one food choice
    foodOption = input("Choose a food option to be served to your guests (1, 2, or 3): ")
    if foodOption.isdigit() == False or int(foodOption) < 1 or int(foodOption) > 3:
        foodOption = input("Please select an option within a range of 1 to 3: ")
    guests = int(guests)

    # stores the total burgers that will be served based on what each guest will receive
    totalConsumedBurgers = 0
    # the total burger packages the provider will purchase
    burgerPackages = 0

    # stores the total buns that will be served based on what each guest will receive  
    totalConsumedBuns  = 0
    # the total buns packages the provider will purchase
    bunsPackages = 0

    # stores the total meat patties that will be served based on what each guest will receive  
    totalConsumedMeatball = 0
    # the total meatball packages the provider will purchase
    meatballPackages = 0

    ## stores the total sandwiches that will be served based on what each guest will receive  
    totalConsumedSandwiches = 0
     # the total chicken packages the provider will purchase (considers chicken pattie same price of meat pattie)
    sandwichPackages = 0
    
    outOfPocketCost = 0.0

    # burger
    if foodOption == "1":
        burgerPerGuest = '0'

        # prompts the user for the number of burgers to be served to each guest
        while burgerPerGuest.isdigit() == False or int(burgerPerGuest) <= 0:
            burgerPerGuest  = input("Please enter the estimated number of burgers each guest will consume? (1 and more): ")
        # the actual number of burgers that will be served
        totalConsumedBurgers = guests * int(burgerPerGuest)
        # computes the number of minimum number of burger packages needed
        burgerPackages = math.ceil(totalConsumedBurgers / BURGERS_PER_PACKAGE)

        # the cost of purchasing the minimum number of burger packages
        burgerCost = burgerPackages * PRICE_BURGER_PACKAGE

        # computes the number of buns to be served alongside the burger
        totalConsumedBuns = guests * int(burgerPerGuest)
        # the minimum number of buns packages
        bunsPackages = math.ceil(totalConsumedBuns / BUNS_PER_PACKAGE)
        bunsCost = bunsPackages * PRICE_BUNS_PACKAGE
        
        # the amount spent on the burger and the bus
        outOfPocketCost += burgerCost + bunsCost
    # sandwich
    elif foodOption == "2":
        sandwichPerGuest = '0'
        # prompts the user for the number of sandwiches to be served to each guest
        while sandwichPerGuest.isdigit() == False or int(sandwichPerGuest) <= 0:
            sandwichPerGuest  = input("Please enter the estimated number of sandwiches each guest will consume? (1 and more): ")
        # the actual number of sandwiches that will be served
        totalConsumedSandwiches = int(guests) * int(sandwichPerGuest)
        # computes the number of minimum number of meat patties packages needed
        sandwichPackages = math.ceil(totalConsumedSandwiches / PATTIES_PER_PACKAGE)

         # the cost of purchasing the minimum number of meat patties packages
        sandwichPrice = sandwichPackages * PRICE_PATTIES_PACKAGE

         # computes the number of buns to be served alongside the burger
        totalConsumedBuns = int(guests) * int(sandwichPerGuest)
        # the minimum number of buns packages
        bunsPackages = math.ceil(totalConsumedBuns / BUNS_PER_PACKAGE)
        # the cost of purchasing the buns will be the total number of buns packages * the Price per bun package
        bunsCost = bunsPackages * PRICE_BUNS_PACKAGE

         # the amount spent on the burger and the bus
        outOfPocketCost += sandwichPrice + bunsCost
    # meatball
    else:
        meatBallsPerGuest = '0'
        # prompts the user for the number of meatballs to be served to each guest
        while meatBallsPerGuest.isdigit() == False or int(meatBallsPerGuest) <= 0:
            meatBallsPerGuest  = input("Please enter the estimated number of meatballs each guest will consume? (1 and more): ")
        # the actual number of meat balls that will be served   
        totalConsumedMeatball = int(guests) * int(meatBallsPerGuest)
        # computes the minimum number of meat patties packages needed  
        meatballPackages = math.ceil(totalConsumedMeatball / PATTIES_PER_PACKAGE)

        # the cost of purchasing the minimum number of meat patties packages
        meatballPrice = meatballPackages * PRICE_PATTIES_PACKAGE

        # computes the number of buns to be served alongside the burger
        totalConsumedBuns = int(guests) * int(meatBallsPerGuest)
        # the minimum number of buns packages
        bunsPackages = math.ceil(totalConsumedBuns / BUNS_PER_PACKAGE)
        # the cost of purchasing the buns will be the total number of buns packages * the Price per bun package
        bunsCost = bunsPackages * PRICE_BUNS_PACKAGE
        
        outOfPocketCost += meatballPrice + bunsCost
        
    print()
    print("You have the option to add additional add-ons to servings (e.g chips or sodas)")
    
    #handle add-on chips
    addChips = input("Do you want to add chips to the servings? (Y/N): ").upper()
    # validates the user's input
    while addChips not in ["Y", "YES", "N", "NO", "NONE"]:
        addChips = input("Please enter Y for yes and N for no: ").upper()
        
    totalChipBags = 0
    if addChips == "Y" or addChips == "YES":
        chipPerGuest = input("How many chip bags per guest do you estimate to serve? ")
        # ensures the user inputs an integer > 0
        while chipPerGuest.isdigit() == False or int(chipPerGuest)<=0:
            chipPerGuest = input("Please enter a whole number greater than 0: ")
        # computes the total chip bags served to all the guests
        totalChipBags = int(chipPerGuest) * int(guests)
        
    chipsCost = totalChipBags * PRICE_PER_CHIP_BAGS
    
    outOfPocketCost += chipsCost
    
    # soda handling
    addSoda = input("Do you want to add soda to the servings? (Y/N): ").upper()
    # validates the user's input
    while addSoda not in ["Y", "YES", "N", "NO", "NONE"]:
        addSoda = input("Please enter Y for yes and N for no: ").upper()
        
    totalCans = 0
    if addSoda == "Y" or addSoda == "YES":
        sodaPerGuest = input("Enter the estimated number of soda cans per guest: ")
        # ensures the user inputs an integer > 0
        while sodaPerGuest.isdigit() == False or int(sodaPerGuest) <=0:
            sodaPerGuest = input("Please enter a whole number greater than 0: ")
        totalCans = int(guests) * int(sodaPerGuest)
    sodaPrice =  totalCans * PRICE_PER_SODA_CAN
    outOfPocketCost += sodaPrice
         
    # prompts the user for the length of the cookout
    duration = input("How long will the cookouts be? ")
    while duration.isdigit() == False or int(duration) < 3:
        duration = input(f"The minimum hosting hours is {MIN_HOURS}. Please enter a whole number 3 and above: ")
    
    
    print("**************************** TOTAL REQUIREMENTS **************************************************")
    # displays the total requirements for the cookout
    print(f"Food items required: {burgerPackages} packages of burgers, {sandwichPackages} packages of chicken patties (sandwich), {meatballPackages} packages of meat patties, {bunsPackages} packages of buns, {totalChipBags} bags of chips, and {totalCans} cans of soda")

    # the total burgers that will be purchased based on the packages of burgers required
    totalPurchasedBurgers = burgerPackages * BURGERS_PER_PACKAGE
    # the left over burger will be total burgers in each purchased package - the actually consumed burgers
    leftOverBurgers = totalPurchasedBurgers - totalConsumedBurgers
    
    # the total sandwiches that will be purchased based on the packages of meat patties required
    totalPurchasedSandwiches = sandwichPackages * PATTIES_PER_PACKAGE
    # the left over sandwich(meat pattie) will be total purchased sandwich in each purchased package - the actually consumed sandwiches
    leftOversandwich = totalPurchasedSandwiches - totalConsumedSandwiches

    # the total meatballs that will be purchased based on the packages of meat patties required
    totalPurchasedMeatballs = meatballPackages * PATTIES_PER_PACKAGE
    # the left over meatballs(meat patties) will be total meatballs in each purchased package - the actually consumed meatballs
    leftOverMeatball = totalPurchasedMeatballs - totalConsumedMeatball

    # the total buns that will be purchased based on the packages of buns required 
    totalPurchasedBuns = bunsPackages * BUNS_PER_PACKAGE
    leftOverBuns = totalPurchasedBuns - totalConsumedBuns

    # the hosting cost is 75 * duration of the cookout
    hostingCost = HOSTING_PRICE_PER_HOUR * int(duration)

    # the amount the customer will pay is 125 % of cost of purchasing the food requirements + the hosting cost
    overallCost = 1.25 * outOfPocketCost + hostingCost

    profit = overallCost - outOfPocketCost
    
    # display costs
    print(f"Leftovers:  {leftOverBurgers} burgers, {leftOversandwich} sandwiches, {leftOverMeatball} meatballs, and {leftOverBuns} buns ")
    print(f"Out of pocket cost: ${outOfPocketCost:.2f}")
    print(f"Catering cost: ${overallCost:.2f}")
    print(f"Profit: ${profit:.2f}")

    # ask the customer if they want to estimate other cookouts
    otherEstimates = input("Do you want to calculate another cook out estimates? (Y/N): ").upper()
    while otherEstimates not in ["Y", "YES", "N", "NO", "NONE"]:
        otherEstimates = input("Please enter Y for yes and N for no: ").upper()
    if otherEstimates == "N" or otherEstimates == "NO" or otherEstimates == "NONE":
        print("Thank you for using this program")
        print("I hope your occasions goes well")
        print("Bye!!  Have a great day.")
        break
    else:
        print()
    

```
